```python
# quantile 
q1 = combineddata['temp'].quantile(.25)
q3 = combineddata['temp'].quantile(.75)
```

```python
def find_outliers(combineddata,cols):
    q1 = combineddata[cols].quantile(0.25)
    q3 = combineddata[cols].quantile(0.75)
    iqr = q3 - q1
    min_r = q1 - 1.5 * iqr
    max_r = q3 + 1.5 * iqr
    outliers_indices = combineddata.index[(combineddata[cols] < min_r) | (combineddata[cols] > max_r)]
    return outliers_indices
```

```python
# find outliers
find_outliers(combineddata,'temp')
```
